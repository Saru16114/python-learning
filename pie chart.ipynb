```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values=[3000,2000,500,450,1000]
exp_labels=['home rent','food','internet','clz fee','other']
```

```python
plt.pie(exp_values,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_values,labels=exp_labels,radius=2)
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_values,labels=exp_labels,radius=2,autopct='%f%%',shadow =True )
plt.show()
```
